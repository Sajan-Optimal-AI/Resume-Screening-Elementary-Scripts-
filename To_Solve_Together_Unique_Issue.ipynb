```python
import spacy
nlp = spacy.load('en')

Input_text = '''Completed BE Computer Science in 2010 at LJCET. Completed ME Computer Science in 2014 at NICHE .'''
tokens = nlp(Input_text)
sentences = []
for sent in tokens.sents:
  sent=str(sent)
  sentences.append(sent)
```

```python
sentences
```

```python
import csv

with open('/content/drive/MyDrive/csv/educational_major.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    Keywords = [val for sublist in data for val in sublist]

Keywords
```

```python
Text_Specialization = {}

for string in sentences:
    for word in Keywords:
        if word in string:
            Text_Specialization[string] = word
            break

print(Text_Specialization )
```
